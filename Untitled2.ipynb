```python
# Imputing missing values for numerical columns using simple imputer
numeric_columns = ['zip', 'lat','long', 'city_pop', 'unix_time','merch_lat', 'merch_long', 'is_fraud']
categorical_columns = ['merchant','category','first','last','gender','street','city','state','job','dob','trans_num']

numeric_imputer =SimpleImputer(strategy='mean')
data_train_imputed = pd.DataFrame(numeric_imputer.fit_transform(data_train[numeric_columns]))
data_test_imputed = pd.DataFrame(numeric_imputer.fit_transform(data_test[numeric_columns]))
```

```python
data_train_imputed.isnull().sum()
```

```python
data_test_imputed.isnull().sum()
```

```python
data_train.isnull().sum()
```

```python
# handling missing values for categorical columns
categorical_imputer = SimpleImputer(strategy='most_frequent')
data_train[categorical_columns] = categorical_imputer.fit_transform(data_train[categorical_columns])# for train dataset
data_test[categorical_columns] = categorical_imputer.fit_transform(data_test[categorical_columns])# for test dataset

```

```python
data_train.isnull().sum()
```

```python
data_test.isnull().sum()
```

#### Changing the given categorical values to numerical representation.

```python
data_train = pd.get_dummies(data_train, columns=categorical_columns, drop_first=True, sparse=True)
```
